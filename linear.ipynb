```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("ass_dataset.csv")
df.head()
```

```python
df.head(20)
```

```python
df.describe()
```

```python
df.plot(x='x',y='y')
plt.xlabel("X =axis")
plt.ylabel("Y-axis")
plt.title("Relation b/w X and Y")
plt.show()
```

```python
plt.bar(df['x'], df['y'])
plt.xlabel("X =axis")
plt.ylabel("Y-axis")
plt.title("Relation b/w X and Y")
plt.show()
```

```python
plt.scatter(df['x'], df['y'])
plt.xlabel("X =axis")
plt.ylabel("Y-axis")
plt.title("Relation b/w X and Y")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
a=df[['x']]
b=df['y']
```

```python
model=LinearRegression()
model.fit(a,b)
```

```python
b_pred = model.predict(a)
plt.scatter(df['x'],df['y'])
```

```python
b_pred = model.predict(a)
plt.scatter(df['x'],df['y'])
plt.plot(df['x'], b_pred, color='red', label='Regression line')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Relation b/w X and Y")
plt.show()
```
